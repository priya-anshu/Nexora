# EDA — Supermarket Sales

Quick exploratory checks for the transaction-level sales dataset used in daily revenue forecasting.

```python
import pandas as pd

sales = pd.read_csv('../data/supermarket_sales.csv')
sales.head()
```

```python
sales['Date'] = pd.to_datetime(sales['Date'])
daily_revenue = sales.groupby('Date', as_index=True)['Total'].sum().sort_index()
daily_revenue.describe()
```
